```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

#metricas
from sklearn import metrics
from sklearn.model_selection import GridSearchCV, cross_val_score,cross_validate,learning_curve
#preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler,MinMaxScaler

#modelos de clasificación
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.linear_model import LogisticRegression

#modelos de regresión
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_excel('Breast_Cancer_Gene.xlsx')
data = data.dropna(how='any', axis=0)
data.reset_index(level=0,inplace=True,drop=True)
```

```python
mask = data['death_from_cancer'].str.fullmatch('Died of Other Causes')
index = data['death_from_cancer'][mask].index.to_list()
data = data.drop(index,axis=0)
```

```python
encoder = LabelEncoder()
encoder_surgery = encoder.fit(data['type_of_breast_surgery'])
encoder_cancer_type_detailed = encoder.fit(data['cancer_type_detailed'])
encoder_cellularity = encoder.fit(data['cellularity'])


type_of_breast_surgery_encoded = encoder_surgery.fit_transform(data['type_of_breast_surgery'])
cancer_type_detailed_encoded = encoder_cancer_type_detailed.fit_transform(data['cancer_type_detailed'])
cellularity_encoded = encoder_cellularity.fit_transform(data['cellularity'])
```

```python
encoder_y = encoder.fit(data['death_from_cancer'])
y_encoded = encoder_y.fit_transform(data['death_from_cancer'])
y = np.array(y_encoded)

x1 = type_of_breast_surgery_encoded
x2 = cancer_type_detailed_encoded
x3 = cellularity_encoded
x4 = data['chemotherapy']
x5 = data['neoplasm_histologic_grade']
x6 = data['mutation_count']
x7 = data['radio_therapy']
x8 = data['tumor_size']

X = np.array([x1,x2,x3,x4,x5,x6,x7,x8]).reshape(-1,8)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.25, random_state=14)
```

***
___

# GridsearchCV

```python
knn_param_grid = {'n_neighbors':[1,5,10,50]}
knn = KNeighborsClassifier()

knn_grid = GridSearchCV(knn,knn_param_grid).fit(X_train,y_train)
print('El mejor score es: ', knn_grid.best_score_)
print('Con los parámetros: ',knn_grid.best_params_)
```

# Cross_validation

```python
scores = ['accuracy']
cross_val = cross_validate(knn_grid, X,y, scoring=scores)
cross_val
```

```python
cross_score = cross_val_score(knn_grid, X,y, cv=5)
cross_score
```

<h1><center> Métricas de evaluación para modelos de Clasificación </center></h1>

# Matriz de confusión
para una clasificación binaria tendríamos:

> Applied machine learning in Python - Kevin. C

### Ecactitud - accuracy

la exactitud es el procentaje de etiquetas estimadas correctamente

 $$accuracy = \frac{TP+TN}{TP+TN+FP+FN}$$

### Error de clasificación - classification error
el error de lcasificación es el procentaje de etiquetas estimadas incorrectamente

$$ce = 1-accuracy$$

### Recall o tasa de verdaderos positivos (TPR)
que porcentaje de los valores verdaderos positivos fue correctamente clasificado?

$$Recall = \frac{TP}{TP+FN}$$

### Tasa de falsos positivos (TNR)
que porcentaje de los valores verdaderos negativos fue correctamente clasificado?
$$TNR = \frac{TN}{TN+FP}$$

### Precisión 
¿Que porcentaje de solo los valores positivos fue correctamente clasificado? 

$$Precision = \frac{TP}{TP+FP}$$

### F1 score

$$F_{1} = 2*\frac{precision*recall}{precision+recall} = \frac{2*TP}{2*TP +FN+FP}$$

# Precision - Recall trade off

Suponga que usted quiere predecir si una documento esta bien clasificado o no, para eso usted usa una métrica de evaluación orientada hacia la precisión dado que quiere predecir la mayor cantidad de verdaderos positivos, pero, si por el contrario usted al estimar el resultado de una prueba diagnósitca sobre el cáncer usted quiere reducir lo máximo posible los verdaderos negativos. 

1. Recall 

Modelos de detección de enfermedades o de clasificación de delitos

2. Precision

Estimación del precio de las acciones o de clasificación de noticias

# Evaluación Multi-clase

```python
data = pd.read_excel('Breast_Cancer_Gene.xlsx')
data = data.dropna(how='any', axis=0)
data.reset_index(level=0,inplace=True,drop=True)

encoder_y = encoder.fit(data['death_from_cancer'])
y_encoded = encoder_y.fit_transform(data['death_from_cancer'])
y = np.array(y_encoded)

encoder = LabelEncoder()
encoder_surgery = encoder.fit(data['type_of_breast_surgery'])
encoder_cancer_type_detailed = encoder.fit(data['cancer_type_detailed'])
encoder_cellularity = encoder.fit(data['cellularity'])


type_of_breast_surgery_encoded = encoder_surgery.fit_transform(data['type_of_breast_surgery'])
cancer_type_detailed_encoded = encoder_cancer_type_detailed.fit_transform(data['cancer_type_detailed'])
cellularity_encoded = encoder_cellularity.fit_transform(data['cellularity'])

x1 = type_of_breast_surgery_encoded
x2 = cancer_type_detailed_encoded
x3 = cellularity_encoded
x4 = data['chemotherapy']
x5 = data['neoplasm_histologic_grade']
x6 = data['mutation_count']
x7 = data['radio_therapy']
x8 = data['tumor_size']

X = np.array([x1,x2,x3,x4,x5,x6,x7,x8]).reshape(-1,8)
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.25, random_state=14)

knn_param_grid = {'n_neighbors':[1,5,10,50]}
knn = KNeighborsClassifier()

knn_grid = GridSearchCV(knn,knn_param_grid).fit(X_train,y_train)
print('El mejor score es: ', knn_grid.best_score_)
print('Con los parámetros: ',knn_grid.best_params_)

y_predict = knn_grid.predict(X_test).reshape(-1,1)
y_true = y_test.reshape(-1,1)

cm = pd.DataFrame(metrics.confusion_matrix(y_true,y_predict))
sns.heatmap(cm,annot=True,cmap='summer',fmt='1g')
plt.show()
```

# Coeficiente de determinación $R^{2}$

Determina la "bondad" de ajuste del modelo, y dice que tanto de la variabilidad de la variable objetivo es explicada por el modelo

$$R^{2} = \frac{\sum(\hat{y_{t}}-\bar{y})^{2}}{\sum(y_{t}-\bar{y})^{2}} = \frac{SEC}{STC}$$

```python
metrics.r2_score(y_test,y_predict)
```

# Error absoluto medio

mide la diferencia media en valor absoluto entre los valores objetivo y los valores predichos 

$$MAE = \frac{1}{n}\sum |y_{t}-\hat{y_{t}}|$$

```python
metrics.mean_absolute_error(y_test,y_predict)
```

# Error cuadrático medio
mide la diferencia al cuadrado entre los valores objetivo y los predichos 

$$MSE = \frac{1}{n}\sum (y_{t}-\hat{y_{t}})^{2}$$

```python
metrics.mean_squared_error(y_test,y_predict)
```

# Mediana del error absoluto 

en lugar de usar la media usa la mediana de las estimaciónes siendo más robusto con valores atípicos

```python
metrics.median_absolute_error(y_test,y_predict)
```
