$$ ventas = \beta_{0}+ \beta_{1}(TV) + u$$

### ¿Para que sirve un modelo de Regresión lineal?
En machine learning la regresión lineal es un tipo de aprendizaje supervisado, donde se quiere predecir el resultado de una variable "objetivo" dependiendo de la variación de las variables "predictoras", en pocas palabras,  "cuánto  puede cambiar $y$ si $X$ cambia"

  + $\beta_{i}$: Se conocen como estimadores del modelo
  + $u$: se conoce como término de error o pertrubación

# Mínimos cuadrados ordinarios

el nombre MCO se refiere al modelo estadístico cuyos estimadores intentan minimizar la suma de los **residuales** elevados al cuadrado 

<h1><center> $\sum_{i=1}^{n}\ \hat u_{i}^{2} = \sum_{i=1}^{n}\ (y_{i}-\hat\beta_{0}-\hat\beta_{1}x_{i})^{2}$ </center></h1>

### Supuestos del modelo de regresión lineal simple

1. los estimadores son lineales
2. los estimadores son insesgados
3. los estimadores minimizan la suma de los errores al cuadrado


4. el valor esperado del término de error u es 0
5. la varianza del error es constante 
6. el término de error distribuye normal

```python
# 1 Escoger las variables
y = np.array(ventas['Sales']).reshape(-1,1)
X = np.array(ventas['TV']).reshape(-1,1)

plt.scatter(X,y)
```

```python
# 2. conjunto de entrenamiento y conjunto de prueba
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=14)

# 3. Entrenamiento del modelo
reg_lin = LinearRegression(fit_intercept=True).fit(X_train,y_train)

# 4. Predicción
reg_lin_predict_ = reg_lin.predict(X_test)

# 5. Evaluación

reg_train_score = reg_lin.score(X_train,y_train)
reg_test_score = reg_lin.score(X_test,y_test)

reg_r2 = r2_score(y_test,reg_lin_predict_)

print("train set score: ", reg_train_score,"\n",
     "test set score: ", reg_test_score,"\n",
     "r2_score", reg_r2)
```

A diferencia del modelo K-NN, los algortimos de regresión no tienen hiperparámetros que se pueda modificar, sin embargo, nosotros podemos realizar algunas transformaciónes con el fin de mejorar la eficiencia de nuestro estimador

## Transformación logarítimica

Realizar transformaciónes sobre nuestras variables nos permite mejorar la eficiencia de nuestro estimador, estas pueden ser la la transformación logarítmica o la potenciación.

```python
y = np.array(ventas['Sales'])
X = np.array(ventas['TV'].apply(np.log)).reshape(-1,1)

# 2. 
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=14)
```

```python
# 3.
reg_linlog = LinearRegression().fit(X_train,y_train)

# 4.
reg_linlog_predict_ = reg_linlog.predict(X_test)

# 5.
reg_linlog_train_score = reg_linlog.score(X_train,y_train)
reg_linlog_test_score = reg_linlog.score(X_test,y_test)

reg_linlog_r2 = r2_score(y_test,reg_linlog_predict_)

print("Score train test: ",reg_linlog_train_score,"\n",
     "Score tes test: ",reg_linlog_test_score,"\n",
     "Score r2_score: ",reg_linlog_r2,"\n")

# Plot
beta_0 = reg_linlog.intercept_
beta_1 = reg_linlog.coef_

fig = plt.figure()

plt.scatter(X,y)
plt.plot(X,beta_0+beta_1*X,color = 'r')
plt.title('r2_score: {:0.2f}'.format(reg_linlog_r2))
plt.show()
```

```python
y = np.array(ventas['Sales'])
X = np.array(ventas['TV'])

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=14)

def pot_1(x):
    y = x**1
    return(y)

transformaciones = np.array([[pot_1,pot_1,np.log,np.log],[pot_1,np.log,np.log,pot_1]]).reshape(4,2)
modelo = ['lin-lin','log-log','lin-log','log-lin']

fig = plt.figure(figsize=(10,7))
subplot = np.arange(221,225,1)

for (ix,i),j,m in zip(enumerate(transformaciones),subplot,modelo):
    
    # 2
    y_train2 = np.array(pd.Series(y_train).apply(transformaciones[ix,0])).reshape(-1,1)
    X_train2 = np.array(pd.Series(X_train).apply(transformaciones[ix,1])).reshape(-1,1)
    
    y_test2 = np.array(pd.Series(y_test).apply(transformaciones[ix,0])).reshape(-1,1)
    X_test2 = np.array(pd.Series(X_test).apply(transformaciones[ix,1])).reshape(-1,1)
    
    # 3
    linear_regresion = LinearRegression().fit(X_train2,y_train2)
    # 4
    linear_regresion_predict_ = linear_regresion.predict(X_test2)
    beta_0 = linear_regresion.intercept_
    beta_1 = linear_regresion.coef_.reshape(-1)

    # 5. 
    r2 = r2_score(y_test2,linear_regresion_predict_)
    
    # Plot
    
    plt.subplot(j)
    plt.scatter(X_train2,y_train2)
    plt.plot(X_train2,beta_0+beta_1*X_train2,color='r')
    plt.legend(['predicciones','data'])
    plt.title('r2_score: {:0.2f} modelo {}'.format(r2,m))
    fig.tight_layout()
plt.show()
```

***
___

Teniendo los mismos datos para todos los modelos vemos que la métrica de evaluación ``r2_score`` cambia dependiendo de la tansformación; obteniendo el mayor puntaje con un modelo log-log, sin embargo, la interpretación del resultado también cambia, siendo esta:

 + **nivel-nivel**
 
 
$ \Delta y = \hat\beta_{1} \Delta x $  

Una variación de 1 **unidad** en $x$ genera  una variación de $\hat\beta_{1}$ **unidades** en $y$

 + **log-nivel**:
 
$ ln(\Delta y) = \hat\beta_{1} \Delta x $ 

Una variación de 1 **unidad** en $x$ genera una variación de $(\hat\beta_{1} * 100)$ **porciento** en $y$

 + **nivel-log**:
 
$ \Delta y = \hat\beta_{1} \Delta ln(x) $

Una variación de 1 **porciento** en $x$ genera una variación de $(\frac{\hat\beta_{1}}{100})$ **unidades** en $y$

 + **log-log**:
 
$ \Delta ln(y) = \hat\beta_{1} \Delta ln(x) $

Una variación de 1 **porciento** en $x$ genera una variación de $\hat\beta_{1}$ **porciento** en $y$

<h1><center>REGRESIÓN LINEAL MÚLTIPLE</center></h1>

## El problema de multicolinealidad exacta

Del modelo de regresión lineal simple sabemos que:
    
<h1> $$y_{i} = \beta_{0} + \beta_{1}x_{i} + u_{i}$$ 

despejando obtenemos:

<h1> $$y_{i} - \beta_{0}+\beta_{1}x_{i} =  u_{i}$$

***
___

puesto que el objetivo del modelo de MCO es minimizar $\sum_{i=1}^{n} u_{i}^{2}$, reescribiendo obtenemos:

<h1> $$min \sum_{i=1}^{n} u_{i}^{2} = min \sum_{i=1}^{n} (y_{i} -\beta_{0}+\beta_{1}x_{i})^{2}$$

Derivando con respecto a $\beta_{1}$ y $\beta_{2}$ e igualando a 0 obtenemos:

\begin{equation}
\begin{split}
    \sum u_{i} &= \sum y_{i} = n\beta{1}+\beta_{2}\sum x_{i}\\
    \sum u_{i}x_{i} &= \sum y_{i}x_{i} = \beta_{i}\sum x_{i} - \beta_{2}\sum x_{i}^{2}
\end{split}
\end{equation}

reescribiendo el último miembro de cada ecuación en su forma matricial obtenemos:


\begin{equation}
\underbrace{\begin{bmatrix} 
T   &   \sum x_{i}\\
\sum x_{i} & \sum x_{i}^{2}
\end{bmatrix}}_{\mathbb{A}}
\overbrace{\begin{bmatrix}
\beta_{0}\\
\beta_{1}
\end{bmatrix}}^{\mathbb{B}}
=
\underbrace{\begin{bmatrix} 
\sum y_{i}\\
\sum y_{i}x_{i}
\end{bmatrix}}_{\mathbb{Y}}
\end{equation}





***
___

La matriz $\mathbb{A}$ es $\mathbb{X}^{T}\mathbb{X}$ y despejando $\mathbb{B}$

<h1> $$\hat{\mathbb{B}} = (\mathbb{X}^{T}\mathbb{X})^{-1}\mathbb{X}^{T}\mathbb{Y}$$

Esta última ecuación implica que la matriz $(\mathbb{X}^{T}\mathbb{X})^{-1})$ es no singular, o lo que es lo mismo, la matriz es de rango completo, es decir, que todas las filas y columnas deben ser linealmente independientes.

si la mztriz no es de de rango completo su determinante es 0 y no se podrá invertir, por ende, los estimadores son indeterminandos y el problema no tendría solución.

Al Modelar regresiones lineales, (y otras), siempre debemos prestar atención al coeficiente de correlación de nuestras variables, un coeficiente de correlación muy alto puede implicar problemas de multicolinealidad exacta que impidan la solución de nuestro modelo.
